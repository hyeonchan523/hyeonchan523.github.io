# Travel distance Optimization

```python
import pyforest
sns.set_style('darkgrid')
```

```python
x = np.linspace(0.1,300,300)
fatigue = x*0.01
fun = np.log(x+1)
```

```python
plt.figure(figsize=(8,5))
happiness = -fatigue
sns.lineplot(x,happiness)
plt.axvline(x[(np.argmax(happiness))],alpha = 0.5,color = 'red')
plt.ylabel('Happiness')
plt.xlabel('Distance')
plt.savefig('../img/optimization_only_fat.png')
```

```python
plt.figure(figsize=(8,5))
happiness = fun
sns.lineplot(x,happiness)
plt.axvline(x[(np.argmax(happiness))],alpha = 0.5,color = 'red')
plt.ylabel('Happiness')
plt.xlabel('Distance')
plt.savefig('../img/optimization_only_hap.png')
```

```python
plt.figure(figsize=(8,5))
happiness = fun - fatigue
sns.lineplot(x,happiness)
plt.ylabel('Happiness')
plt.xlabel('Distance')
plt.axvline(x[(np.argmax(happiness))],alpha = 0.5,color = 'red')
plt.savefig('../img/optimization_moderate.png')
```

```python
plt.figure(figsize=(8,5))
happiness = 1.1*fun - 0.9*fatigue
sns.lineplot(x,happiness)
plt.ylabel('Happiness')
plt.xlabel('Distance')
plt.axvline(x[(np.argmax(happiness))],alpha = 0.5,color = 'red')
plt.savefig('../img/optimization_moderate_1.png')
```

# Machine Learning 

```python
from sklearn.linear_model import LinearRegression
```

```python
x = np.linspace(0,5)
X = x.reshape(-1,1)
y = x+ np.random.normal(0,0.5,50)
```

```python
plt.figure(figsize = (8,5))
sns.scatterplot(x,y)
plt.savefig('../img/optimization_linearRegression')
```

```python
plt.figure(figsize = (8,5))

sns.scatterplot(x,y)
for i in range(10):
    y_line = np.random.uniform(0.1,2)*(x - 2.5 + np.random.uniform(-0.5,0.5)) +3
    sns.lineplot(x,y_line)
plt.ylim(0,6)    
plt.savefig('../img/optimization_random')
```

```python
reg = LinearRegression()
reg.fit(X,y)
y_pred = reg.coef_ * x + reg.intercept_
```

```python
plt.figure(figsize = (8,5))
sns.scatterplot(x,y)
sns.lineplot(x,y_pred,color = 'r')
plt.savefig('../img/optimization_linearRegression_w_pred')
```
