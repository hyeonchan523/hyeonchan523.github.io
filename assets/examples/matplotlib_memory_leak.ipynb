```python
import os
import psutil

def _check_usage_of_cpu_and_memory():
    
    pid = os.getpid()
    py  = psutil.Process(pid)
        
    memory_usage  = round(py.memory_info()[0] /2.**30, 2)
    return memory_usage
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
memory_ = np.zeros(100)
dummy_data = np.random.rand(100,1000,1000,3)
for i in range(100):
    plt.imshow(dummy_data[i,:])
    memory_use = _check_usage_of_cpu_and_memory()
    memory_[i] = memory_use

```

```python
plt.figure(figsize = (12,10))
plt.plot(memory_)
plt.title('Memory uses', fontsize = 25)
plt.xlabel('iteration', fontsize = 15)
plt.ylabel('memory useage', fontsize = 15)
plt.savefig('../img/memory_leak_y.png', dpi = 200)
```

```python
memory_2 = np.zeros(100)
plt.close('all')
plt.clf()
dummy_data = np.random.rand(100,1000,1000,3)
for i in range(100):
    plt.imshow(dummy_data[i,:])
    plt.clf()
    plt.close('all')
    memory_use = _check_usage_of_cpu_and_memory()
    memory_2[i] = memory_use

```

```python
plt.figure(figsize = (12,10))
plt.plot(memory_2[10:])
plt.title('Memory uses', fontsize = 25)
plt.xlabel('iteration', fontsize = 15)
plt.ylabel('memory useage', fontsize = 15)
plt.savefig('../img/memory_leak_n.png', dpi = 200)
```
