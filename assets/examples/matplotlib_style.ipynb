```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,2)
y = x**2 + np.random.randn(50)
```

# Seaborn Style

```python
import seaborn as sns
styles = ['darkgrid','whitegrid','dark','white','ticks']
plt.figure(figsize = (10,20))
for i,style in enumerate(styles):
    plt.subplot(5,1,i+1)
    sns.set_style(style)
    plt.plot(x,y)
    plt.title(style)
plt.savefig('../img/seaborn_style.png', bbox_inches = 'tight')
```

```python
mpl_styles = plt.style.available
plt.figure(figsize = (15,60))
for i,style in enumerate(mpl_styles):
    plt.subplot(13,2,i+1)
    plt.style.use(style)
    plt.plot(x,y)
    plt.title(style,fontsize = 15,color = 'k')
plt.savefig('../img/mpl_style.png',bbox_inches = 'tight')
```
